```python
import matplotlib.pyplot as plt

months = [  1,2, 3, 4, ]
sales = [2222,5555,7777,8888]

plt.plot(months,sales ,color = "black" , linestyle = "--" , linewidth = 2 , marker = "o" , label = "first program")
plt.title ("partice program")
plt.xlabel("months")
plt.ylabel("sales pr months")
plt.xlim(0.5 , 4.5 )
plt.ylim(1000,10000)
plt.grid(color = "gray" , linestyle = ':', linewidth = 1)
plt.xticks([1,2,3,4],["m1","m2","m3","m4"])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [20,10,15,25]

plt.subplot(1 , 2 , 1 )
plt.plot(x, y , color = "b" , linewidth = 2 , marker  = "o")

plt.subplot(1 , 2 , 2 )
plt.bar(x, y , color = "y")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["Amit", "Sneha", "Ravi", "Priya"]
marks = [85, 92, 78, 88]

# First subplot (1 row, 2 cols, position 1)
plt.subplot(1, 2, 1)
plt.bar(students, marks, color="skyblue")
plt.title("Marks - Bar Chart")
plt.ylabel("Scores")

# Second subplot (1 row, 2 cols, position 2)
plt.subplot(1, 2, 2)
plt.plot(students, marks, marker="o", color="orange")
plt.title("Marks - Line Chart")
plt.ylabel("Scores")

plt.tight_layout()

plt.savefig  ('program.png' , dpi=300 , bbox_inches = "tight")
plt.show()
```
